```python
from ltp import LTP
ltp = LTP()
```

```python
output = ltp.pipeline(["他叫汤姆去拿外衣。"], tasks=["cws", "pos", "ner", "srl", "dep", "sdp"])
print(output.cws)  #分词
print(output.pos)
print(output.sdp)
```
